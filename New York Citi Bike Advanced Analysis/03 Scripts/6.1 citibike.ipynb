# 03. Data wrangling

```python
# Checking for missing values
df.isnull().sum()
```

### Only birth year has missing values. We can't drop them, because there are too many. Those rows will contain valuable information needed for our analysis.

```python
# Dropping unnecessary column
df = df.drop(columns = ['trip_id'])
```

```python
df
```

```python
# Converting gender from int to string
df['gender'] = df['gender'].astype('str')
```

```python
df.info()
```

```python
# Checking and creating subset for anyone older than 95
df_sub = df.loc[df['birth_year'] < 1917]
```

```python
df_sub
```

### Since Citi Bike's age requirement is 16 years to open up an account and 95 years old is pushing the limit, anyone older than that can be viewed as an outlier. So years 1899, 1900 etc cannot be changed to 1999 or 2000 etc thinking that it was a mistake, because you have to be 16 to open an account as this information was for September 2013. So these 23 entries will be dropped

```python
# Dropping above rows
df = df.drop(df[df['birth_year'] < 1917].index)
```

```python
df
```

```python
df.describe()
```

```python
# Replacing all NaN values in birth_year column with mean of 1976.0
df['birth_year'].fillna(1976.0, inplace = True)
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
df
```

```python
df.info()
```

```python
# Checking the days of the week
df['weekday'].unique()
```

```python
# Checking gender values
df['gender'].value_counts(dropna = False)
```

```python
# Creating a function to re-categorize gender
def gender_category(row):
    if row['gender'] == '1':
        return 'Male'
    elif row['gender'] == '2':
        return 'Female'
    else: return 'Unknown'
```

```python
# Applying the function
df['gender_cat'] = df.apply(gender_category, axis = 1)
```

```python
df
```

```python
# Changing start_time column from string to datetime
df['start_time'] = df['start_time'].astype('datetime64[ns]')
```

```python
# Changing end_time column from string to datetime
df['end_time'] = df['end_time'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
# Dropping gender column
df = df.drop(columns = ['gender'])
```

```python
df
```

```python
# Renaming gender_cat column
df.rename(columns = {'gender_cat' : 'gender'}, inplace = True)
```

```python
df
```

```python
df.info()
```

```python
# Changing start_station_id column from int to string
df['start_station_id'] = df['start_station_id'].astype('str')
```

```python
# Changing end_station_id column from int to string
df['end_station_id'] = df['end_station_id'].astype('str')
```

```python
df.info()
```

```python
# Changing birth_year column from float to int
df['birth_year'] = df['birth_year'].astype('int64')
```

```python
df
```
